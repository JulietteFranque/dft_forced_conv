```python
import os
os.chdir('../')
from forced_convection_dft.forward_models import ForwardModelDft
from dft import dft_models, heat_transfer_coefficients
os.chdir('jupyter_notebooks')
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_points = 500
time_vector = np.linspace(0, n_points + 1, n_points)
incident_heat_flux = np.ones(n_points) * 20e3
velocity =  np.ones(n_points) * 3
incident_heat_flux[:100] = 0
incident_heat_flux[-100:] = 0


diameter_tc = 1.4e-3
density_tc = 8665
heat_capacity_tc = 485.6 
emissivity_tc=0.8
alpha_tc=0.8

length_dft_side =  0.0016
length_insulation = .019

temperature_initial = 20 + 273
```

```python
time_vector[1] - time_vector[0]
```

```python
forward_dft = ForwardModelDft(number_of_nodes=10, time_vector=time_vector, insulation_thickness=length_insulation, plate_thickness=length_dft_side, incident_heat_flux=incident_heat_flux,
                 characteristic_velocity=velocity, initial_temperature=temperature_initial, temperature_ambient=temperature_initial, temperature_surroundings=temperature_initial)
```

```python
temperature_front, temperature_back, temperature_insulation, coefficient_front, coefficient_back = forward_dft.run()
```
