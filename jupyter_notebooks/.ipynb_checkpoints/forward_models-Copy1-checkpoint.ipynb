```python
import os
os.chdir('../')
from forced_convection_dft.forward_models import ForwardModelDft, ForwardModelThermocouple
from forced_convection_dft.correlations import PlateCorrelations, SphereCorrelations
from dft import dft_models, heat_transfer_coefficients
os.chdir('jupyter_notebooks')
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_points = 500
time_vector = np.linspace(0, 500, n_points)
velocity =  np.ones(n_points) * 5


diameter_tc = 1.4e-3
density_tc = 8665
heat_capacity_tc = 485.6 
emissivity_tc=0.8
alpha_tc=0.8

length_dft_side =  0.0016
length_insulation = .019

temperature_initial = 21 + 273
```

```python
fluxes = np.linspace(.0001, .0005, 5)
#fluxes = np.linspace(5, 150, 5)
mean_percentage_errors = []
for flux in fluxes:
    incident_heat_flux = time_vector**2 * flux * 1e3 #W/m2
    #incident_heat_flux = np.ones(n_points) * flux * 1e3
    #incident_heat_flux[time_vector < 100] = 0
    #incident_heat_flux[time_vector > 400] = 0
    forward_dft = ForwardModelDft(number_of_nodes=10, time_vector=time_vector, insulation_thickness=length_insulation, plate_thickness=length_dft_side, incident_heat_flux=incident_heat_flux,
                 characteristic_velocity=velocity, initial_temperature=temperature_initial, temperature_ambient=temperature_initial, temperature_surroundings=temperature_initial)
    temperature_front, temperature_back, temperature_insulation, coefficient_front, coefficient_back = forward_dft.run()
    coefficient_front_nat = heat_transfer_coefficients.natural_convection(
            (temperature_front + temperature_initial).reshape(-1, 1) / 2, Kelvin=True,
           T_infty=temperature_initial).custom(0.65, 0.25).flatten()
    #print(coefficient_front)
    inverse_model_nat_conv = dft_models.one_dim_conduction(T_f=temperature_front, T_b=temperature_back, time=time_vector, h_f=coefficient_front_nat, h_b=coefficient_back, model='one_d_conduction', Kelvin=True, T_inf=temperature_initial, T_sur=temperature_initial)
    inverse_model_nat_forced = dft_models.one_dim_conduction(T_f=temperature_front, T_b=temperature_back, time=time_vector, h_f=coefficient_front, h_b=coefficient_back, model='one_d_conduction', Kelvin=True, T_inf=temperature_initial, T_sur=temperature_initial)
    errors = np.abs((incident_heat_flux/1e3 - inverse_model_nat_conv.q_inc) / (incident_heat_flux/1e3) * 100)
    mean_error = np.mean(errors[np.isfinite(errors)])
    mean_percentage_errors.append(mean_error)
    plt.plot(time_vector, incident_heat_flux/1e3, label='true', color='orange')
    plt.plot(time_vector, inverse_model_nat_conv.q_inc, label='assuming nat', color='red')
    plt.plot(time_vector, inverse_model_nat_forced.q_inc, label='using forced h', color='dodgerblue')
    plt.ylabel('heat flux (kW/m2)')
    plt.legend()
    plt.figure()
```

```python
#incident_heat_flux = time_vector ** 2 * .001 * 1e3
incident_heat_flux = np.ones(n_points) * 10 * 1e3
#incident_heat_flux[time_vector < 100] = 0
#incident_heat_flux[time_vector > 400] = 0

plt.plot(time_vector, incident_heat_flux / 1e3)
forward_dft = ForwardModelDft(number_of_nodes=10, time_vector=time_vector, insulation_thickness=length_insulation, plate_thickness=length_dft_side, incident_heat_flux=incident_heat_flux,
                 characteristic_velocity=velocity, initial_temperature=temperature_initial, temperature_ambient=temperature_initial, temperature_surroundings=temperature_initial)
temperature_front, temperature_back, temperature_insulation, coefficient_front, coefficient_back = forward_dft.run()


forward_tc = ForwardModelThermocouple(time_vector, incident_heat_flux, velocity, temperature_initial,
                 temperature_initial, temperature_initial, emissivity_tc, alpha_tc, diameter_tc, heat_capacity_tc,
                 density_tc)
temperature_tc = forward_tc.run()
```

```python
def find_thermocouple_q_inc(h_tc, thermocouple_temperature, ambient_temperature, temperature_surroundings, diameter, emissivity, absorptivity, heat_capacity, density):
    convective_heat_flux = h_tc * (thermocouple_temperature - ambient_temperature)
    area = 4 * np.pi * (diameter / 2) ** 2
    volume = 4 / 3 * np.pi * (diameter / 2) ** 3
    emitted_heat_flux = emissivity * 5.67e-8 * (
            thermocouple_temperature ** 4 - temperature_surroundings ** 4)
    energy_stored = (density * heat_capacity * volume) * np.gradient(thermocouple_temperature, time_vector)
    incident_heat_flux = (energy_stored/area + convective_heat_flux + emitted_heat_flux) / absorptivity
    return incident_heat_flux / 1e3
```

```python
def minimize_error(velocity, ambient_temperature, temperature_tc, temperature_front_dft, temperature_back_dft, plate_thickness, diameter, emissivity, absorptivity, heat_capacity, density):
    h_front_dft = PlateCorrelations(surface_temperature=temperature_front_dft, ambient_temperature=ambient_temperature,characteristic_length=plate_thickness).find_forced_convection_heat_transfer_coefficient(velocity)
    h_back_dft = heat_transfer_coefficients.natural_convection(
            (temperature_back + temperature_initial).reshape(-1, 1) / 2, Kelvin=True,
           T_infty=temperature_initial).custom(0.65, 0.25).flatten()
    h_tc = SphereCorrelations(
                surface_temperature=temperature_tc,
                ambient_temperature=ambient_temperature,
                characteristic_length=diameter).find_forced_convection_heat_transfer_coefficient(velocity)
    q_inc_plate = dft_models.one_dim_conduction(T_f=temperature_front, T_b=temperature_back, time=time_vector, h_f=coefficient_front, h_b=coefficient_back, model='one_d_conduction', Kelvin=True, T_inf=temperature_initial, T_sur=temperature_initial).q_inc
    q_inc_tc = find_thermocouple_q_inc(h_tc, temperature_tc, ambient_temperature, ambient_temperature, diameter, emissivity, absorptivity, heat_capacity, density)
    #plt.plot(q_inc_plate, label='plate')
    #plt.plot(q_inc_tc, label='tc')
    #plt.legend()
    return np.abs(q_inc_plate - q_inc_tc).sum()
    
    
```

```python
minimize_error(velocity, temperature_initial, temperature_tc, temperature_front, temperature_back, length_dft_side, diameter_tc, emissivity_tc, alpha_tc, heat_capacity_tc, density_tc)


```

```python
from scipy.optimize import minimize
```

```python
x0 = np.ones(time_vector.shape[0])
res = minimize(lambda x: minimize_error(x, temperature_initial, temperature_tc, temperature_front, temperature_back, length_dft_side, diameter_tc, emissivity_tc, alpha_tc, heat_capacity_tc, density_tc ), x0, method='powell',
               options={'ftol': 1, 'disp': True}, bounds=list((((0, np.inf), ) * time_vector.shape[0])))


```
